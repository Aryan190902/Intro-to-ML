# Face Detection - OpenCV

```
!pip install opencv-python
```

```
# importing opencv
import cv2
```

```
group_photo = plt.imread("/content/group-photo.png",format='PNG')
```

```
group_photo.shape
```

```
plt.imshow(group_photo)
```

```
model = cv2.CascadeClassifier("/content/haarcascade_frontalface_alt.xml")
```

```
faces = model.detectMultiScale(group_photo, 1.2 )
```

```
faces.shape
```

```
faces
```

```
# drawn all the rectangles
for face in faces:
  x,y,w,h = face
  group_photo = cv2.rectangle(group_photo, (x, y), (x+w, y+h), (255,0,0), 3 )
```

```
plt.imshow(group_photo)
```
